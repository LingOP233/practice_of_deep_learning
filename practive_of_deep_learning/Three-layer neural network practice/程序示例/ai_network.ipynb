```python
import numpy as np

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # 
    return exp_x / np.sum(exp_x)

# 
x = np.linspace(-5, 5, 100)
y = softmax(x)

print("Softmax :", y)
print(":", np.sum(y))  #  1

plt.plot(x, y, label="ReLU Function", color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Softmax Activation Function")
plt.legend()
plt.grid()
plt.show()
```

```python
# 
num_examples = len(X)

# 
nn_input_dim = 2

# 2 one-hot 
nn_output_dim = 2

# 
epsilon = 0.01        # learning rate
reg_lambda = 0.01     # L2 
```

```python
# 
def calculate_loss(model):
    # 
    W1, b1 = model['W1'], model['b1']
    W2, b2 = model['W2'], model['b2']

    # 
    z1 = X.dot(W1) + b1                   #   
    a1 = np.tanh(z1)                      # tanh
    z2 = a1.dot(W2) + b2                  #   
    exp_scores = np.exp(z2)               #  e^score
    probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)  # softmax 

    # 
    correct_logprobs = -np.log(probs[range(num_examples), y])
    data_loss = np.sum(correct_logprobs)

    #  L2 
    data_loss += (reg_lambda / 2) * (
        np.sum(np.square(W1)) + np.sum(np.square(W2))
    )

    # 
    return data_loss / num_examples
```

```python
#  x0  1
def predict(model, x):
    # 
    W1, b1 = model['W1'], model['b1']
    W2, b2 = model['W2'], model['b2']

    # 
    z1 = x.dot(W1) + b1          #   
    a1 = np.tanh(z1)             # tanh
    z2 = a1.dot(W2) + b2         #   
    exp_scores = np.exp(z2)      # 
    probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)  # softmax 

    # 
    return np.argmax(probs, axis=1)
```

```python
# 
# 
# - nn_hdim
# - num_passes
# - print_loss 1000 
def build_model(nn_hdim, num_passes=30000, print_loss=False):
    np.random.seed(0)

    #  +  0
    W1 = np.random.randn(nn_input_dim, nn_hdim) / np.sqrt(nn_input_dim)
    b1 = np.zeros((1, nn_hdim))
    W2 = np.random.randn(nn_hdim, nn_output_dim) / np.sqrt(nn_hdim)
    b2 = np.zeros((1, nn_output_dim))

    model = {}

    # Batch GD
    for i in range(num_passes):
        # --------  --------
        z1 = X.dot(W1) + b1               #   
        a1 = np.tanh(z1)                  # tanh
        z2 = a1.dot(W2) + b2              #   
        exp_scores = np.exp(z2)           # softmax 
        probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)  # softmax 

        # --------  --------
        delta3 = probs
        delta3[range(num_examples), y] -= 1         #  - 
        dW2 = a1.T.dot(delta3)                      # 
        db2 = np.sum(delta3, axis=0, keepdims=True) # 

        delta2 = delta3.dot(W2.T) * (1 - np.power(a1, 2))  # tanh 1 - tanh(x)
        dW1 = X.T.dot(delta2)                             # 
        db1 = np.sum(delta2, axis=0)                      # 

        # -------- L2--------
        dW2 += reg_lambda * W2
        dW1 += reg_lambda * W1

        # -------- --------
        W1 -= epsilon * dW1
        b1 -= epsilon * db1
        W2 -= epsilon * dW2
        b2 -= epsilon * db2

        # 
        model = {'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2}

        #  1000 
        if print_loss and i % 1000 == 0:
            print(f" {i} {calculate_loss(model):.6f}")

    return model
```

```python
#  3 
model = build_model(nn_hdim=3, print_loss=True)
plt.figure(figsize=(8, 6))
# 
plot_decision_boundary(lambda x: predict(model, x))
plt.title("Decision Boundary for hidden layer size 3")
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid(True)
plt.savefig("ai_net_img_03.png",dpi=300)
plt.show()
```
